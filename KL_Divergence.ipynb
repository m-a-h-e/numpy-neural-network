```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
```

```python
d = 0.05
va = np.arange(d, 6.0 + d, d)
ma = np.arange(-1.5, 1.5, d)
```

```python
vmesh, mmesh = np.meshgrid(va, ma)

v = vmesh.ravel()
m = mmesh.ravel()

z = v - np.log(v) - 1 + m*m

zmesh = z.reshape(vmesh.shape)
```

```python
fig = plt.figure()
ax = Axes3D(fig)

ax.plot_surface(vmesh, mmesh, zmesh);
ax.set_xlabel('variance')
ax.set_ylabel('mean');
ax.set_zlabel('KL Divergence');
```

```python
plt.pcolor(vmesh, mmesh, zmesh, cmap='coolwarm');
```

```python
v = np.arange(0.01, 6.0, 0.02)
y = v - np.log(v) - 1.0          # variance part (mean = 0) of KL Divergence => minimum at variance = 1.0

plt.plot(v, y);
```
